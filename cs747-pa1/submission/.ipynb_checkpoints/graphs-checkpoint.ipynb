```python
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
output1 = "./outputDataT1.txt"
```

```python
df1 = pd.read_csv(output1, sep=", ", engine="python", names=['instance', 'algo', 'seed', 'ep', 'horizon', 'reg'])
```

```python
df1['horizon'] = df1.horizon.apply(math.log)
df1.head()
```

```python
reg_df1 = df1.groupby(['instance', 'algo', 'horizon'])['reg'].mean()
reg_df1.head()
```

```python
reg_df1['../instances/i-2.txt']['thompson-sampling']
```

```python
plt.figure()
reg_df1['../instances/i-3.txt']['epsilon-greedy'].plot()
reg_df1['../instances/i-3.txt']['ucb'].plot()
reg_df1['../instances/i-3.txt']['kl-ucb'].plot()
reg_df1['../instances/i-3.txt']['thompson-sampling'].plot(grid=True)
plt.legend(("epsilon-greedy", "ucb", "kl-ucb", 'thompson-sampling'))
```

```python
plt.figure()
reg_df1['../instances/i-2.txt']['epsilon-greedy'].plot()
reg_df1['../instances/i-2.txt']['ucb'].plot()
reg_df1['../instances/i-2.txt']['kl-ucb'].plot()
reg_df1['../instances/i-2.txt']['thompson-sampling'].plot(grid=True)
plt.legend(("epsilon-greedy", "ucb", "kl-ucb", 'thompson-sampling'))
```

```python
plt.figure()
reg_df1['../instances/i-1.txt']['epsilon-greedy'].plot()
reg_df1['../instances/i-1.txt']['ucb'].plot()
reg_df1['../instances/i-1.txt']['kl-ucb'].plot()
reg_df1['../instances/i-1.txt']['thompson-sampling'].plot(grid=True)
plt.legend(("epsilon-greedy", "ucb", "kl-ucb", 'thompson-sampling'))
```
